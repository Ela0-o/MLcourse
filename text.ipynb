```python
import numpy as np
from sklearn.datasets import fetch_20newsgroups
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline

data = fetch_20newsgroups()
```

```python
X_train, y_train = fetch_20newsgroups(subset='train', return_X_y=True, random_state=42)
X_test, y_test = fetch_20newsgroups(subset='test', return_X_y=True, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier as KNN
mod = KNN()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
from sklearn.model_selection import GridSearchCV

parameter_dict = {
    'tfidf__min_df': [1, 5, 10],
    'tfidf__max_df': [0.1, 0.3, 1],
    'model__n_neighbors': [2, 5],
    'model__weights': ['uniform', 'distance'],
    'model__metric': ['cosine', 'minkowski']
}
# Combine TfidfVectorizer with LinearRegression
pipeline = Pipeline([
    ('tfidf', TfidfVectorizer()),
    ('model', KNN())
])
gscv = GridSearchCV(estimator=pipeline,param_grid=parameter_dict, scoring='neg_mean_squared_error', cv=3, verbose=100)

gscv.fit(X_train, y_train)
```

```python
best_min = gscv.best_params_['tfidf__min_df']
best_max = gscv.best_params_['tfidf__max_df']
print(best_min,best_max)
```
